Group BY

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)

sns.barplot(x='Size', y='Qty', data=qty)
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, x='ship-state', color='orange')
plt.xticks(rotation=90)
plt.xlabel('state')
plt.ylabel('count')
plt.show()
```

VereficaÃ§Ã£o de vendas por estados.

```python
state_top = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,5))
sns.countplot(data=df[df['ship-state'].isin(state_top.index)], x='ship-state')
plt.xticks(rotation=45)
plt.title('Sales by States')
plt.xlabel('state')
plt.ylabel('count')
plt.show()
```

No grÃ¡fico acima que o estado com a maior taxa de vendas na amazon Ã© o estado de 'Maharashtra', e o produto com maior nÃºmero de vendas Ã© T-shirt.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category', color='green') 
plt.title('Sales of Category')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data=df, x='State',y='Sales', color='orange')
plt.title('Sales of State')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='Sub-Category', color='purple')
plt.title('Sales of Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

```python

ax = sns.countplot(data=df, x='Sub-Category', color='purple')
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure(figsize=(15, 6))
ax = sns.barplot(data=df, x='State',y='Profit', color='gray')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.title('Profit of State')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=df, x='State',y='Sales', color='red')
plt.xticks(rotation=90)
plt.title('Sales of State')
plt.show()
```

```python
top_10 = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
top_10.plot(kind='bar', figsize=(10, 6), color='purple')
plt.title('Top 10 States by Sales') 
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```
